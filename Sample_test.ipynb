```python
import pandas as pd
df = pd.read_csv('/content/sample_data/missing_data_practice.csv')
```

**Finding the missing values**

```python

df.isna()
```

```python
df.sum().isna()
```

```python
#checking for the datatypes
print(df.dtypes)
```

```python
#Displaying the columns having missing values using a loop
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        continue
    else:
        print(f"Column '{col}' has non-numeric data.")
```

```python

# Convert a specific column to numeric, setting invalid parsing to NaN
# Step 1: Define column types properly

numeric_cols = ['age', 'marks', 'salary']
date_cols = ['joining_date']

# Step 2: Convert numeric columns
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Step 3: Convert date columns
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

```

```python
df.isna().sum()
```

**Handling the missing values**

```python
df.dropna()
```

```python
df.dropna(how ='all')
```

```python
df.dropna(thresh=6)
```

```python
df['name']=df['name'].fillna('n/a')
df['marks'] = df['marks'].fillna(0)
df['age']=df['age'].fillna(0)
df['department'] = df['department'].fillna('n/a')
df['joining_date'] = df['joining_date'].fillna('n/a')
df['salary']=df['salary'].fillna(0)
```

```python
df.isna().sum()
```

```python
df
```

**Retrieve the information to analyze**

```python
#Display all employees whose salary is greater than 50,000
df[df['salary']>50000]
```

```python
#Display employees who belong to the IT department AND have marks greater than 80
df[(df['department']=='IT') & (df['marks']>80)]
```

```python
#Display all rows where joining_date is missing.
df[df['joining_date']=='n/a']
```

```python
#Display rows where age is less than 18 OR greater than 60 OR is missing.
df[(df['age']<18)|
    (df['age']>60)|
    (df['age']==0)
  ]
```
